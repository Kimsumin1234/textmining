```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
# 음식 리뷰 예제 텍스트 데이터
corpus = [
    "샐러드 맛이 깔끔해서 좋아요.",
    "주문한 음식이 늦게 와서 배고팠어요.",
    "피자가 맛있어요.",
    "라면이 너무 짜게 나왔어요.",
    "배달된 치킨이 맛있어요.",
    "메뉴 설명보다 양이 적어요.",
    "주문한 음식이 다른 것으로 와서 다시 주문했어요.",
    "맛은 좋은데 양이 좀 적은 편이에요.",
    "떡볶이의 매운맛이 적절해요.",
    "초밥이 신선해서 좋았어요.",
    "빵이 부드럽고 맛있네요!",
    "스테이크가 너무 질겼어요.",
    "샐러드의 드레싱이 너무 강했어요.",
    "나초의 치즈소스가 맛있었어요.",
    "파스타가 약간 덜 익은 것 같아요.",
    "커피가 너무 진해서 물을 더 추가했어요.",
    "디저트는 달지 않아서 좋았습니다.",
    "음료가 너무 달아서 다음엔 당도 조절할게요.",
    "피자 도우가 너무 두꺼워서 먹기 힘들었어요.",
    "햄버거의 빵이 너무 딱딱해요.",
    "감자튀김이 너무 기름져서 느끼했어요.",
    "치즈케이크가 아주 부드럽고 맛있어요."
]
```

## 전처리
### 형태소 분석기를 통한 조사 구두점 제거
* https://github.com/hyunwoongko/pecab
* 조사, 구두점 제거
* 품사태그표 :
    * https://konlpy.org/ko/v0.6.0/morph/
    * https://docs.google.com/spreadsheets/d/1OGAjUvalBuX-oZvZ_-9tEfYD2gQe7hTGsgUpiiBSXI8/edit#gid=0

### 텍스트 전처리 전 벡터화

```python
# 벡터화 단계

cvect = CountVectorizer()
dtm = cvect.fit_transform(corpus)
dtm
```

```python
# 행렬 조회
print(dtm.toarray())

# 단어장 확인
# 문제점 : 아무조건 없이 그냥 벡터화를 시키면, 같은단어로 개수를 셀수있는것도
# '샐러드' '샐러드의' , '피자' '피자가' 와 같이 같은의미의 단어 들인데 세분화 되서 나뉘게된다
print(cvect.get_feature_names_out())
```

```python
# pecab 을 사용해서 문제점 해결하기
# pure python 으로 되어 있는 한국어 형태소 분석기

from pecab import PeCab

pecab = PeCab()
```

```python
pecab.pos(corpus[-1])
```

```python
# 제거할 품사(조사) 작업하기

for text,pos in pecab.pos(corpus[-1]):
  # J 로 시작하는 품사 제외하기
  if not pos.startswith('J'):
    print(text,pos)
```

```python
len(corpus)
```

```python
# 위 개념을 적용한 함수 작성
# 조사(J로시작)와 구둣점(느낌표,SF) 제거

corpus_pecab = []

for corpus_text in corpus:
  join_text = []
  for text,pos in pecab.pos(corpus_text,drop_space=False):
    if (not pos.startswith('J')) & (pos not in ['SF']):
      # print(text,pos)
      join_text.append(text)
  corpus_pecab.append(''.join(join_text))

corpus_pecab
```

### 텍스트 전처리 후 벡터화

```python
# 75 => 67

cvect = CountVectorizer()
dtm = cvect.fit_transform(corpus_pecab)
dtm
```

```python
cvect.get_feature_names_out()
```

```python
df_dtm = pd.DataFrame(dtm.toarray(),columns=cvect.get_feature_names_out())
df_dtm
```

## 군집화

* https://scikit-learn.org/stable/modules/clustering.html#k-means
* https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html
- 각각의 클러스터에 중심과 가까운 걸 모은것

```python
# K-means 군집화
from sklearn.cluster import KMeans

# 임의로 클러스터 생성 개수를 지정
num_clusters = 7

# max_iter=300 (기본값) : 클러스터의 중심점을 찾기 위해서 300번 반복해줘
kmeans = KMeans(n_clusters=num_clusters, n_init='auto', random_state=42)
kmeans
```

```python
# KMeans 기본형태 : kmeans.fit(df_dtm).predict(df_dtm)
# KMeans 는 transform() 을 안쓰고 predict() 을 사용

# KMeans 가 fit() 완료후 labels_ 추출이 가능함 == .predict(df_dtm)
# lables_ 로 추출하면 되서 굳이 .predict(df_dtm) 안하고 fit() 에서 끝내도 된다
clusters = kmeans.fit(df_dtm)
clusters
```

```python
kmeans.labels_
```

## 군집화 결과
* 같은 내용의 텍스트끼리 군집화가 되었는지 확인

```python
# label 번호의 의미 : 같은 label 번호로 corpus 가 묶어서 나온다

df = pd.DataFrame()
df['corpus'] = corpus
df['label'] = kmeans.labels_
df.sort_values('label')
```

```python
# 군집화 결과

df['label'].value_counts()
```

```python
df_dtm.shape
```

```python
# (22, 67) 를 모두 표현하기에는
# 차원축소
from sklearn.manifold import TSNE

# perplexity= : 이웃의 수 (숫자가 커질수록 간격이 넓혀진다)
# tsne = TSNE(n_components=2, perplexity=5, random_state=42)
tsne = TSNE(n_components=2, perplexity=10, random_state=42)
tsne_result = tsne.fit_transform(df_dtm)
tsne_result # x,y 좌표로 표현해줌
```

```python
df_tsne = pd.DataFrame(tsne_result, columns=['x','y'])
df_tsne['corpus'] = corpus
df_tsne['label'] = kmeans.labels_
df_tsne
```

```python
sns.scatterplot(data=df_tsne, x='x', y='y', hue='label',palette='Set1')
plt.text(x=df_tsne.loc[0,'x'],y=df_tsne.loc[0,'y'],s=df_tsne.loc[0,'corpus'])
```

```python
# 전체 텍스트 표현하기

plt.figure(figsize=(15,10))
sns.scatterplot(data=df_tsne, x='x', y='y', hue='label',palette='Set1')
for idx in df_tsne.index:
  plt.text(x=df_tsne.loc[idx,'x'],y=df_tsne.loc[idx,'y']+2,s=df_tsne.loc[idx,'corpus'])
```

### TfidfVectorizer 벡터화를 사용해서 해보기

```python
# 벡터화

tfidfvect = TfidfVectorizer()
dtm = tfidfvect.fit_transform(corpus_pecab)

df_dtm = pd.DataFrame(dtm.toarray(),columns=tfidfvect.get_feature_names_out())
df_dtm
```

```python
# 군집화

kmeans = KMeans(n_clusters=num_clusters, n_init='auto', random_state=42)
clusters = kmeans.fit(df_dtm)
```

```python
# 군집화 결과
df = pd.DataFrame()
df['corpus'] = corpus
df['label'] = kmeans.labels_
df.sort_values('label')
```

```python
# 군집화 결과

df['label'].value_counts()
```

```python
# 보기쉽게 데이터 프레임 형태로 만들기

df_tsne = pd.DataFrame(tsne_result, columns=['x','y'])
df_tsne['corpus'] = corpus
df_tsne['label'] = kmeans.labels_
df_tsne
```

```python
# 차트로 시각화

plt.figure(figsize=(15,10))
sns.scatterplot(data=df_tsne, x='x', y='y', hue='label',palette='Set1')
for idx in df_tsne.index:
  plt.text(x=df_tsne.loc[idx,'x'],y=df_tsne.loc[idx,'y']+2,s=df_tsne.loc[idx,'corpus'])
```
